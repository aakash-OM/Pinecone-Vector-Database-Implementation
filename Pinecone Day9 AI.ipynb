## Create Embeddings for each of the Text Chunk

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embedding, index_name=index_name)
```

```python
#querry related to uploded pdf
query = "Education"
```

```python
#apply similarity search
docs = docsearch.similarity_search(query) #docsearch is an obj that stores all the  embedding
```

```python
docs
```

```python
#calling openAPI (to convert these numeric terms into words)
llm = OpenAI()
```

```python
#RetrievalQA: class (used for question answering)
#from_chain_type : method
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "Education"
```

```python
qa.run(query)
#getting highly relatabe answer for the question(query) because of the similarity score(score based on cosine-similarity).
```

```python
#to implement Q n A system:

import sys
while True:
  user_input = input(f"Input Prompt: ")
  if user_input == 'exit':
    print('Exiting')
    sys.exit()
  if user_input == '':
    continue
  result = qa({'query': user_input})
  print(f"Answer: {result['result']}")
```
